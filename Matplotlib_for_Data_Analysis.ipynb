```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_excel("expense3.xlsx")
df = pd.DataFrame(data)

# Group by "Category" and sum "Amount"
group_by = df.groupby("Category").agg({"Amount": "sum"})

# Print grouped data
print(group_by)

# Convert categorical index to numeric
x_values = range(len(group_by))

# Plot with step
plt.step(x_values, group_by["Amount"], where='pre',marker = "o")

# Label x-axis with category names
plt.xticks(x_values, group_by.index, rotation=45)

# Show plot
plt.show()
```

SAVE A CHART USING - Matplotlib

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5,6,7]

NOP1 = [5,10,30,20,35,60,80]
NOP2 = [10,20,30,75,80,90,120]
NOP3 = [8,30,50,60,70,90,100]

plt.stackplot(days,NOP1,NOP2,NOP3,colors = ["pink","hotpink","magenta"],labels = ["week1","week2","week3"])
plt.legend()
plt.savefig("stackplot.png",facecolor = "white",pad_inches = 0.9,bbox_inches = "tight")
plt.show()
```
